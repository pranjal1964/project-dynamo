Since the diagnosis_result column is in text , we need to convert it to binary so that it can be fed into the algorithm.

```python
df=data
```

```python
#using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df.drop('diagnosis_result',axis=1))
#removing the last column of the dataframe as this table will be used for the feature matrix
scaled_features=scaler.transform(df.drop('diagnosis_result',axis=1))
new_data=pd.DataFrame(scaled_features,columns=df.columns[:-1])
```

```python
data.head()
```

```python
#feature matrix
X = df.drop(columns=['diagnosis_result'])
```

```python
#target matrix
y=df['diagnosis_result']
```

```python
# Importing the train_test_split function from the scikit-learn library
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset into training and testing sets, with 80% of the data used for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# random value of n_neighbors, we will find a better value of k later.
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
# Use the trained KNN model to make predictions on the test data
pred=knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Calculate the accuracy of the model by comparing the true labels (y_test) with the predicted labels (y_pred)
accuracy = accuracy_score(y_test, pred)
print("Accuracy:", accuracy)
```

```python
# Generated and print a classification report comparing the true labels (y_test) with the predicted labels (pred)
print(classification_report(y_test,pred))
```
