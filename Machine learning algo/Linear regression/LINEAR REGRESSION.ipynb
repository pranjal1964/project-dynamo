# Linear Regression Algorithm

```python

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import accuracy_score
import pandas as pd
```

1)  sklearn.linear_model import LinearRegression- This Library is a machine learning library for python. It is the part of scikit-learn(sklearn). Linear Regression class implements the linear regression model using scikit-learn.It gves the tools for linear regression.
 sklearn.
2) model_selection.train_test_split: This library is also part of scikit-learn and is used for splitting datasets into two subsets: one for training the model and the other for testing the model. It randomly splits the dataset into training and testing sets based on the specified test size or train size.
3) sklearn.metrics.mean_absolute_error: This library provides a function to calculate mean absolute error (MAE) between the actual and predicted values. MAE is a measure of the average absolute difference between the actual and predicted values.
4) sklearn.metrics.mean_squared_error: As  this library  provides a function to calculate the mean squared error (MSE) between the actual and predicted values. MSE is a measure of the average squared difference between the actual and predicted values.
5) sklearn.metrics.r2_score: While using this library it gives a function to calculate the coefficient of determination (R-squared) between the actual and predicted values. R-squared is a measure of how well the independent variables explain the variability of the dependent variable.
6) from sklearn.metrics import accuracy_score: This line imports the accuracy_score  from the sklearn.metrics module. This function is used to calculate the accuracy of classification models by comparing the predicted labels with the true labels.

```python

x = df[['km']]

y = df['current price']
```

x = df[['km']] it is defining the features variable which is the independent variables.
y = df['current price'] it is defining the target variable which is the dependent variable.

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

In the above code I have split the data into training and testing sets.

```python
model = LinearRegression()
```

model = LinearRegression() this is used to find  patterns in data to predict an outcome based on the relationship between input variables (like price of a car) and an output variable (like km).

```python
model.fit(x_train, y_train)
```


This line of code trains the linear regression model on the training data, allowing it to learn the relationship between the independent variables (features) and the dependent variable (target) in order to make predictions.

```python
y_pred = model.predict(x_test)
```


This line of code uses the trained linear regression model to predict the target variable (y) values based on the features (x) of the test data.

```python
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```


This code is calculating mean squared error (MSE), mean absolute error (MAE), and R-squared, to assess the performance of the linear regression model on the test data.

```python
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

Above lines are printing the mean square error , absolute error and R-Square score.

```python
import matplotlib.pyplot as plt

# Plotting the actual vs predicted values
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices")
plt.show()
```

This code creates a scatter plot whch is comparing the actual prices (y_test) with the predicted prices (y_pred). The x-axis represents the actual prices, and the y-axis represents the predicted prices. 

```python
import matplotlib.pyplot as plt

# Plotting the actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', label='Actual')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, color='red', label='Predicted')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.legend()
plt.show()
```

This code creates a scatter plot where the actual values are shown along the x-axis and the predicted values along the y-axis. It helps visualize how well the predictions match the actual values.
