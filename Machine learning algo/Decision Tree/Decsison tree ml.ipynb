```python
# Imported of necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
# Loading of the dataset
df= pd.read_csv('C:/Users/Hp/Downloads/apple_quality.csv')
```

```python
# df is used to show the contents of the dataset.
df
```

```python
#The info() function is showing the number of rows and columns, what kind of data is stored in each column (like numbers or text), how many values are missing
df.info()
```

```python
#  removes rows from the DataFrame df that contain missing values (NaN). 
df=df.dropna()
df
```

```python
# Selecting the column containing acidity values from the DataFrame df
df['Acidity']
```

```python
# Converting the 'Acidity' column in the DataFrame to float data type
df['Acidity']=df['Acidity'].astype(float)
```

```python
# Acidity coloumn converted into data type float
df['Acidity']
```

```python
# Removing the 'A_id' column from the DataFrame 
df.drop(columns='A_id', inplace=True)
```

```python
# Assigned the  features X and target variable Y
X = df[['Size', 'Weight', 'Sweetness', 'Crunchiness', 'Juiciness', 'Ripeness',
       'Acidity']]  # Assumed these  features
Y = df['Quality']  # Assumed target variable quality'

# Splited the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# Defining the Decision Tree classifier
clf = DecisionTreeClassifier(random_state=0)
```

```python

# The hyperparameters grid is defined to specify a range of values for each hyperparameter
# This grid will be used in hyperparameter tuning to find the best combination of hyperparameters
param_grid = {
    'criterion': ['gini', 'entropy'], # The criterion for splitting: 'gini' or 'entropy'
    'max_depth': [None, 5, 10, 15, 20], # The maximum depth of the decision tree
    'min_samples_split': [2, 5, 10],  # The minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4] # The minimum number of samples required to be at a leaf node
}
```

```python

# This line of code imports the train_test_split function and the GridSearchCV class from the sklearn.model_selection module.
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
# Create the GridSearchCV object
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
```

```python
# Perform grid search to find the best hyperparameters
grid_search.fit(X_train, y_train)
```

```python
# Print the best hyperparameters found
print("Best hyperparameters:", grid_search.best_params_)
```

```python
# Get the best estimator
best_clf = grid_search.best_estimator_
```

```python
# Predict on the test set using the best classifier
y_pred = best_clf.predict(X_test)
```

```python
# X_train contains the features used for training the machine learning model
X_train
```

```python

# Importing the DecisionTreeClassifier class from the sklearn.tree module
# Importing accuracy_score, classification_report, and confusion_matrix functions from sklearn.metrics module
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create a decision tree classifier
model = DecisionTreeClassifier(random_state=0)

# Train the model on the training set
model.fit(X_train, y_train)

```

```python
# Predicting the target variable (Y_pred) using the trained model on the test set (X_test)
Y_pred = model.predict(X_test)
```

```python
# contains the predicted values of the target variable obtained from the trained model.
Y_pred
```

```python
# Importing the confusion_matrix function from the sklearn.metrics module.
from sklearn.metrics import confusion_matrix
```

```python
# Calculating the confusion matrix to evaluate the performance of the classifier.
cm=confusion_matrix(y_test,Y_pred)
cm
```

```python

# Calculating the accuracy score to measure the accuracy of the classifier's predictions.
accuracy = accuracy_score(y_test, Y_pred)
accuracy
```
