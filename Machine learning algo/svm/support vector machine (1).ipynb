```python
# Define features and target variable
X = df.drop(columns=['customer_id', 'churn'])  # Features
y = df['churn']  # Target variable
```

```python
# Initialize a Support Vector Classifier (SVC) with a linear kernel
clf = SVC(kernel='linear')
```

```python
# Split the dataset into training and testing sets
# X_train: Features for training
# X_test: Features for testing
# y_train: Target variable for training
# y_test: Target variable for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python

from sklearn import svm
# Create a Support Vector Machine (SVM) model with a linear kernel
svm_model = svm.SVC(kernel='linear')  
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define numerical and categorical features
numeric_features = ['credit_score', 'age', 'tenure', 'balance', 'products_number', 'estimated_salary']
categorical_features = ['country', 'gender', 'credit_card', 'active_member']

# Create transformers for numerical and categorical features
numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(drop='first')

# Combine transformers for numerical and categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

```

```python
# Apply preprocessing to training and testing sets
X_train_scaled = preprocessor.fit_transform(X_train)
X_test_scaled = preprocessor.transform(X_test)
```

```python
# Create a Support Vector Machine (SVM) model with a linear kernel
svm_model = svm.SVC(kernel='linear')
```

```python
# Trained the SVM model on the scaled training data
svm_model.fit(X_train_scaled, y_train)
```

```python
# Use the trained SVM model to make predictions on the scaled testing data
y_pred = svm_model.predict(X_test_scaled)
```

```python
# Calculating the accuracy of the model by comparing  predicted labels (y_pred) with the true labels (y_test)
accuracy = accuracy_score(y_test, y_pred)
# Generating a classification report, which includes precision, recall, F1-score, and support for each class
report = classification_report(y_test, y_pred) 
```

```python
# Print the accuracy value using formatted string literal
print(f"Accuracy: {accuracy}")
# Print the classification report, including precision, recall, F1-score, and support for each class
print("Classification Report:\n", report)
```
